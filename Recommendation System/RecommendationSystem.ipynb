```python
import pandas as pd
import numpy as np
```

```python
book_df = pd.read_csv('book.csv',encoding='windows-1252')
```

```python
book_df=book_df.iloc[:,1:]
```

```python
book_df=book_df.rename(columns={'User.ID':'user_id','Book.Title':'book_title','Book.Rating':'book_rating'})
```

```python
user_book_df = book_df.pivot_table(index='user_id',
                                 columns='book_title',
                                 values='book_rating').reset_index(drop=True)
```

```python
user_book_df
```

```python
user_book_df.index=book_df.user_id.unique()
```

```python
user_book_df.head()
```

```python
user_book_df.fillna(0,inplace=True)
```

```python
user_book_df.head(5)
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim=1-pairwise_distances(user_book_df,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
```

```python
user_sim_df.head()
```

```python
user_sim_df.index = book_df.user_id.unique()
user_sim_df.columns = book_df.user_id.unique()
```

```python
user_sim_df.head()
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```
